```python
import pandas as pd
import folium
```

```python
df = pd.read_csv('./data/parking.csv', encoding='utf-8')
```

```python
pd.set_option('display.max_columns', 10)
```

```python
df
```

```python
df.columns
```

```python
df = df[['주차장명', '주차장 위치 좌표 위도', '주차장 위치 좌표 경도', '주차 면(주차 가능 차량 수)']]
```

```python
df
```

```python
df.columns = ['명칭', '위도', '경도', '대수']
```

```python
df
```

```python
df = df.dropna(axis=0)
```

```python
df
```

```python
df.head(n=10)
```

```python
parking_map = folium.Map(location=[37.55, 126.98], tile='Stamen Terrain', zoom_start=12)
```

```python
parking_map
```

```python
df
```

```python
for name, lat, lng in zip(df.명칭, df.위도, df.경도):
    folium.Marker([lat, lng],
                  icon=folium.Icon(color='red', icon='info-sign'),
                  popup=name).add_to(parking_map)
```

```python
parking_map.save('./result/parking_map.html')
```
